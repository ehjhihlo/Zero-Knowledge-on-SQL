```python
import pandas as pd
url='https://www.basketball-reference.com/leagues/NBA_2023_per_game.html'
df=pd.read_html(url)[0] 
# df.to_csv('NBA.csv', index=False)
```

```python
df
```

```python
drop_list = []
for i in range(1, len(df)):
    if df['Player'][i] == 'Player':
        drop_list.append(i)    
```

```python
df = df.drop(drop_list)
```

```python
df.to_csv('NBA.csv', index=False)
```
